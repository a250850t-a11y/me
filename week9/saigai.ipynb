```python
import pandas as pd
import matplotlib.pyplot as plt

def load_data(filepath: str) -> pd.DataFrame:
    """CSVファイルを読み込む関数"""
    return pd.read_csv(filepath)

def compute_yearly_counts(df: pd.DataFrame) -> pd.Series:
    """年ごとの災害件数を計算"""
    return df.groupby("Start Year").size()

def plot_yearly_counts(counts: pd.Series) -> None:
    """年ごとの災害件数を折れ線グラフとして描画"""
    plt.figure(figsize=(10, 5))
    plt.plot(counts.index, counts.values, marker='o')
    plt.xlabel("Year")
    plt.ylabel("Disaster Count")
    plt.title("Disasters per Year")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def main():
    df = load_data("emdat.csv")
    counts = compute_yearly_counts(df)
    plot_yearly_counts(counts)

if __name__ == "__main__":
    main()
```
